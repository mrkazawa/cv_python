```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as pl

# import image and display
pil_im = Image.open('resources/empire.jpg')

pl.figure()
pl.imshow(pil_im)
pl.gray()
pl.axis('off')
pl.show()
```

```python
# import images and convert to grayscale image using 'L' arguments
pil_im = Image.open('resources/empire.jpg').convert('L')

pl.imshow(pil_im)
pl.show()
```

```python
# get all images in one directory
import os
filelist = os.listdir('resources')

for infile in filelist:
    outfile = os.path.splitext(infile)[0] + ".jpg"
    print(outfile)
    if infile != outfile:
        try:
            Image.open(infile).save(outfile)
        except IOError:
            print("cannot convert", infile)
```

```python
# creating thumbnail, please notice that the size now different
pil_im.thumbnail((128,128))
pl.imshow(pil_im)
pl.show()
```

```python
# copy and paste region
box = (100,100,400,400)
pil_im = Image.open('resources/empire.jpg').convert('L')
region = pil_im.crop(box)

pl.imshow(region)
pl.show()
```

```python
# create a function to wasily display image
def show_image(src):
    pl.figure()
    pl.imshow(src)
    pl.axis('off')
    pl.show()
```

```python
region = region.transpose(Image.ROTATE_180)
pil_im.paste(region, box)

show_image(pil_im)
```

```python
pil_im = Image.open('resources/empire.jpg').convert('L')

out = pil_im.resize((128,128))
show_image(out)
```

```python
out = pil_im.rotate(45)
show_image(out)
```
