```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
from pylab import *
```

```python
kernel_ed1 = np.array([[1.,0.,-1.],
                       [0.,0.,0.],
                       [-1.,0.,1.]])

kernel_ed2 = np.array([[0.,1.,0.],
                       [1.,-4.,1.],
                       [0.,1.,0.]])

kernel_ed3 = np.array([[-1.,-1.,-1.],
                       [-1.,8.,-1.],
                       [-1.,-1.,-1.]])

kernel_shp = np.array([[0.,-1.,0.],
                       [-1.,5.,-1.],
                       [0.,-1.,0.]])

kernel_bbl = np.array([[1.,1.,1.],
                       [1.,1.,1.],
                       [1.,1.,1.]])/9

kernel_gb1 = np.array([[1.,2.,1.],
                       [2.,4.,2.],
                       [1.,2.,1.]])/16

kernel_gb2 = np.array([[1.,4.,6.,4.,1.],
                       [4.,16.,24.,16.,4.],
                       [6.,24.,36.,24.,6.],
                       [4.,16.,24.,16.,4.],
                       [1.,4.,6.,4.,1.]])/256

kernel_list = [kernel_ed1, kernel_ed2, kernel_ed3, kernel_shp, kernel_bbl, kernel_gb1, kernel_gb2]
title_list = ['Edge Detection 1', 'Edge Detection 2', 'Edge Detection 3', 'Sharpen', 'Box Blur',
              'Gaussian Blur 3 x 3', 'Gaussian Blur 5 x 5']
```

```python
image = array(Image.open('dongseo_gv.jpg'))

plt.rcParams['figure.figsize'] = (16,9)
plt.title('Original Image')
plt.imshow(image)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
output = cv2.filter2D(image, -1, kernel_list[0])

plt.rcParams['figure.figsize'] = (16,9)
plt.title(title_list[0])
plt.imshow(output)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
output = cv2.filter2D(image, -1, kernel_list[1])

plt.rcParams['figure.figsize'] = (16,9)
plt.title(title_list[1])
plt.imshow(output)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
output = cv2.filter2D(image, -1, kernel_list[2])

plt.rcParams['figure.figsize'] = (16,9)
plt.title(title_list[2])
plt.imshow(output)
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(13, 9))

output = cv2.filter2D(image, -1, kernel_list[3])
fig.add_subplot(2,2,1)
plt.title(title_list[3])
plt.imshow(output)
plt.axis('off')

output = cv2.filter2D(image, -1, kernel_list[4])
fig.add_subplot(2,2,2)
plt.title(title_list[4])
plt.imshow(output)
plt.axis('off')

output = cv2.filter2D(image, -1, kernel_list[5])
fig.add_subplot(2,2,3)
plt.title(title_list[5])
plt.imshow(output)
plt.axis('off')

output = cv2.filter2D(image, -1, kernel_list[6])
fig.add_subplot(2,2,4)
plt.title(title_list[6])
plt.imshow(output)
plt.axis('off')

plt.tight_layout()
plt.show()
```
